# Aufgabe 1.2
Importieren Sie den Datenmenge house data.csv als Pandas DataFrame (CSV-Import).
Bestimmen Sie für jedes Attribut
- Mittelwert = `mean`
- Median = `median`
- Varianz = `var`
- Standardabweichung = `std`

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Import dataframe

```python
house_data_raw = pd.read_csv("data/house_data.csv")
house_data_raw
```

## Recalculate squarefoot to squaremeter (Aufgabe 1.4)

```python
house_data = house_data_raw

squarefoot_in_squaremeter = 0.092903
house_data["sqft_living"] = house_data_raw["sqft_living"] * squarefoot_in_squaremeter
house_data["sqft_lot"] = house_data_raw["sqft_lot"] * squarefoot_in_squaremeter

house_data
```

## Describe the imported dataframe
`data.describe()` is used to view basic statistical details like percentile, mean, std etc. of a data frame.

`.apply(lambda s: s.apply(lambda x: format(x, 'g')))` surpress scientifc notation.

```python
house_data.describe().apply(lambda s: s.apply(lambda x: format(x, 'g')))
```

## Printing individual columns

Here's a reminder:
- Mittelwert = `mean`
- Median = `median`
- Varianz = `var`
- Standardabweichung = `std`

### Mittelwert

```python
house_data.mean().round(decimals=3).apply(lambda x: format(x, '3f'))
```

### Median

```python
house_data.median().round(decimals=3).apply(lambda x: format(x, '3f'))
```

### Varianz

```python
house_data.var().round(decimals=3).apply(lambda x: format(x, '3f'))
```

### Standardabweichung

```python
house_data.std().round(decimals=3).apply(lambda x: format(x, '3f'))
```

# Aufgabe 1.5

## Wie teuer ist das Haus mit dem größten Wohnzimmer?

```python
house_data[house_data["sqft_living"] == house_data["sqft_living"].max()]
```

## Wie viele Häuser wurden zwischen 1990 und 2010 gebaut?

```python
house_data[house_data.yr_built.between(1990, 2010)].sort_values(by="yr_built", ascending = True)
```

## Wie teuer ist das renovierte Haus mit dem kleinsten Grundstück?

```python
renovated = house_data[house_data['yr_renovated'] != 0]
renovated[renovated["sqft_lot"] == renovated["sqft_lot"].min()]
```

## Wie viele Häuser haben ein Wohnzimmer, dass min. 70 m² groß ist?

```python
house_data[house_data["sqft_living"] >= 70].sqft_living.count()
```

## In welchem Jahr wurde das günstigste Haus mit 2 Stockwerken gebaut?

```python
two_floors = house_data[house_data["floors"] == 2]
two_floors[two_floors["sqft_living"] == two_floors["sqft_living"].min()]
```

## Welche Häuser haben 3 Stockwerke und einen überdurchschnittlichen Preis obwohl sie eine unterdurchschnittliche Anzahl an Schlafzimmern und Badezimmern haben?

```python
three_floors = house_data[house_data["floors"] == 3]
price_above_avg = three_floors[three_floors["price"] > three_floors["price"].mean()]
bedrooms_below_avg = price_above_avg[price_above_avg["bedrooms"] < price_above_avg["bedrooms"].mean()]
bedrooms_below_avg
```
